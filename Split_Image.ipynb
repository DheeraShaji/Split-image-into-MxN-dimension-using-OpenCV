```python
# libraries required
import numpy as np
from PIL import Image
import os
import cv2
from pathlib import Path
from matplotlib import pyplot as plt
```

```python
#defining paths
source_path = "/content/drive/MyDrive/Computer vision/image.jpeg"
dest_path = "/content/drive/MyDrive/Computer vision/"
```

```python
#function to crop the image into parts

def imgcrop(img_path, m, n):
    file_name = Path(img_path).stem
    im = Image.open(img_path)
    img_width, img_height = im.size
    h = img_height // n
    w = img_width // m
    for i in range(0, n):
        for j in range(0, m):
            b = (j * w, i * h, (j + 1) * w, (i + 1) * h)
            crop_img = im.crop(b)
            #a = a.resize((640,640))
            x=dest_path + file_name + "-patch-" + str(i) + "-" + str(j)+".jpg"
            crop_img.save(x)
```

```python
#function call
imgcrop(source_path,3,3)
```
